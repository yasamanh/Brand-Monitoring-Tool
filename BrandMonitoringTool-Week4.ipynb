## Discovering Stories by Topic Modeling

I give an overview of stories covered in the twitter collection using topic modeling and clustering tweets together as stories. Steps to create clusters of tweets that are less noisy and more informative, including: 

* divided my twitter collection into time slices of one week.
* Noise and common word removal 
* Tweet tokenizing using NLTK
* TFIDF
* Topic modeling(Latent Semantic Analysis) 
* added a probability threshold to further remove the noise tweets in each cluster.
